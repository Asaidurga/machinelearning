```python
import pandas as pd
import numpy as np
#loaad the dataset
dataset=pd.read_csv("/content/priya.csv")
#printing the dataset
dataset
```

```python
#Extracting independent variable(X)and dependent(Y)
X=dataset.iloc[:,:-1].values
Y=dataset.iloc[:,-1].values
print('independent variable(X):',X)
print('dependent variable(Y):',Y)
```

```python
#dealing with missing data
from sklearn.impute import SimpleImputer
#repalcing the missing value transform method
imp=SimpleImputer(missing_values=np.nan,strategy="mean")
X=imp.fit_transform(X)
Y=Y.reshape(-1,1)
Y=imp.fit_transform(Y)
Y=Y.reshape(-1)
print(X)
print(Y)
```

```python
#convert X and Y to DataFrames
X_df=pd.DataFrame(X,columns=dataset.columns[:-1])
Y_df=pd.DataFrame(Y,columns=[dataset.columns[-1]])
#combined X and Y side by side using pandas.concat
combined_Data=pd.concat([X_df,Y_df],axis=1)
#Display the combined data
print("combined data:")
print(combined_Data)
```

```python
#convert the combined_Data to a DataFrame
combined_df=pd.DataFrame(combined_Data,columns=list(dataset.columns))
combined_df.to_excel("/content/priya.csv.xlsx",index=False)
print("combined dataframe:")
print(combined_df.head())
```
